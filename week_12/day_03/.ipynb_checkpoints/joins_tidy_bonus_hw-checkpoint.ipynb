## 1. Read in all 4 credit card transaction datasets and clean up the column names.

```python
card_base = pd.read_csv("data/CardBase.csv")
customer_base = pd.read_csv("data/CustomerBase.csv")
fraud_base = pd.read_csv("data/FraudBase.csv")
transaction_base = pd.read_csv("data/TransactionBase.csv")
```

```python
card_base
```

```python
card_base = card_base.rename(columns = {"Card_Number":"card_number", "Card_Family":"card_family", "Credit_Limit":"credit_limit", "Cust_ID":"cust_id"})
```

```python
customer_base
```

```python
customer_base = customer_base.rename(columns = {"Cust_ID":"cust_id", "Age":"age", "Customer_Segment":"customer_segment","Customer_Vintage_Group":"customer_vintage_group" })
```

```python
fraud_base 
```

```python
fraud_base  = fraud_base.rename(columns={"Transaction_ID":"transaction_id", "Fraud_Flag":"fraud_flag" })
```

```python
transaction_base
```

```python
transaction_base = transaction_base.rename(columns={"Transaction_ID":"transaction_id","Transaction_Date":"transaction_date", "Credit_Card_ID":"card_number", "Transaction_Value":"transaction_value", "Transaction_Segment":"transaction_segment"})
```

## Q2. Join the data containing card details and customer details by customer id, so that all records of card details and any matching records in customer details are kept.

```python
card_base.merge(customer_base, how = 'inner', on = "cust_id")
```

## 3. Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

```python
fraud_base.merge(transaction_base, how = "outer", on = "transaction_id")
```

The number of rows represent the number of transactions

## 6. Unite day, month, and year columns into a column called date using a suitable separator. Then find the date where the most berets were observed.

```python
# Convert day, month, and year columns to appropriate data types
hat_observations['day'] = pd.to_numeric(hat_observations['day'])
hat_observations['month'] = pd.to_numeric(hat_observations['month'])
hat_observations['year'] = pd.to_numeric(hat_observations['year'])
```

```python
hat_observations
```

```python
# Convert integer columns to strings
hat_observations['day'] = hat_observations['day'].astype(str)
hat_observations['month'] = hat_observations['month'].astype(str)
hat_observations['year'] = hat_observations['year'].astype(str)
```

```python
# Combine day, month, and year columns into a new 'date' column
hat_observations['date'] = hat_observations[['day', 'month', 'year']].agg('/'.join, axis=1)
```

```python
hat_observations
```

```python
observation_by_date =hat_observations.groupby('date')['observation_count'].sum()
```

```python
observation_by_date.idxmax()
```
